```python
df_pri['LoanStatus'].value_counts()
```

```python
condition = (df_pri['LoanStatus'] == 'Completed') | (df_pri['LoanStatus'] == 'Defaulted') | (df_pri['LoanStatus'] == 'Chargedoff')
df = df_pri[condition]
def change_to_defaulted(row):    
    if row['LoanStatus'] == 'Chargedoff':        
        return 'Defaulted'    
    else:        
        return row['LoanStatus']    
df_pri['LoanStatus'] = df.apply(change_to_defaulted, axis=1)
df_pri['LoanStatus'].value_counts()
```

### 5B)

```python
categories = {1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'}
```

```python
def categories_class(val):
    if val==1:
        return 'Debt Consolidation'
    elif val==2:
        return 'Home Improvement'
    elif val==3:
        return 'Business'
    elif val==6:
        return 'Auto'
    else:
        return 'Other'
```

```python
condition=df_pri[(df_pri['LoanStatus'] == 'Completed') | (df_pri['LoanStatus'] == 'Defaulted') | (df_pri['LoanStatus'] == 'Chargedoff')]
```

```python
df = condition['ListingCategory (numeric)']
```

```python
df_pri['ListingCategory (numeric)'] = df.apply(categories_class)
```
